## Step 1: Raw Data Mapping

This code file is to read NHANES raw .xpt files, map every variable to its question text and every numeric code to its human-readable meaning, then save everything to one .xlsx.

#### Import Libaries

```python
#pip install pyreadstat
```

```python
import glob
import os
import pandas as pd
import pyreadstat
```

#### Data Path

```python
BASE_DIR = "."
RAW_DIR     = os.path.join(BASE_DIR, "Nhanes_RawData")
MAPPED_XLSX = os.path.join(RAW_DIR, "Nhanes_Mapped.xlsx")  

FILE_TOPICS = {                                       # short name -> friendly label
    "DEMO_L": "Demographics", "DPQ_L": "Depression (PHQ-9)", "ALQ_L": "Alcohol Use",
    "SMQ_L": "Smoking", "SLQ_L": "Sleep", "PAQ_L": "Physical Activity",
    "HSQ_L": "Health Status", "INQ_L": "Income",
}
```

#### Reading xpt files in the Raw Folder,  with data formating (data, metadata). Try utf-8, fall back to latin1.

```python
def read_xpt(path):
    for enc in ("utf-8", "latin1"):
        try:    return pyreadstat.read_xport(path, encoding=enc)
        except Exception: continue
    raise RuntimeError(f"Could not read {path}")

data, metas = {}, {}
for path in sorted(glob.glob(os.path.join(RAW_DIR, "*.xpt"))):
    name = os.path.splitext(os.path.basename(path))[0]      # DPQ_L.xpt -> DPQ_L
    data[name], metas[name] = read_xpt(path)
    print(f"loaded {name:8s} shape={data[name].shape}")
```

#### The value mapping

The value mapping below is transcribed MANUALLY from the NHANES codebook:
https://wwwn.cdc.gov/nchs/nhanes  ->  select the questionnaire (e.g. "DPQ_L").
Each variable has a "Values" table listing every code and its meaning The .xpt file gives us the question TEXT for free (via meta); only these code , meaning pairs must be added by hand from the codebook.

```python
YESNO = {1: "Yes", 2: "No", 7: "Refused", 9: "Don't know"}
PHQ   = {0: "Not at all", 1: "Several days", 2: "More than half the days",
         3: "Nearly every day", 7: "Refused", 9: "Don't know"}

VALUE_LABELS = {
    # DPQ_L : PHQ-9 depression (target)
    "DPQ010": PHQ, "DPQ020": PHQ, "DPQ030": PHQ, "DPQ040": PHQ, "DPQ050": PHQ,
    "DPQ060": PHQ, "DPQ070": PHQ, "DPQ080": PHQ, "DPQ090": PHQ,
    "DPQ100": {0: "Not at all difficult", 1: "Somewhat difficult",
               2: "Very difficult", 3: "Extremely difficult",
               7: "Refused", 9: "Don't know"},
    # DEMO_L : demographics
    "RIDSTATR": {1: "Interviewed only", 2: "Interviewed + MEC examined"},
    "RIAGENDR": {1: "Male", 2: "Female"},
    "RIDRETH1": {1: "Mexican American", 2: "Other Hispanic", 3: "Non-Hispanic White",
                 4: "Non-Hispanic Black", 5: "Other Race - Incl Multi-Racial"},
    "RIDRETH3": {1: "Mexican American", 2: "Other Hispanic", 3: "Non-Hispanic White",
                 4: "Non-Hispanic Black", 6: "Non-Hispanic Asian",
                 7: "Other Race - Incl Multi-Racial"},
    "RIDEXMON": {1: "Nov 1 - Apr 30", 2: "May 1 - Oct 31"},
    "DMQMILIZ": YESNO,
    "DMDBORN4": {1: "Born in 50 US States or DC", 2: "Others",
                 77: "Refused", 99: "Don't know"},
    "DMDEDUC2": {1: "Less than 9th grade", 2: "9-11th grade (no diploma)",
                 3: "High school grad / GED", 4: "Some college or AA degree",
                 5: "College graduate or above", 7: "Refused", 9: "Don't know"},
    "DMDMARTZ": {1: "Married / Living with partner",
                 2: "Widowed / Divorced / Separated", 3: "Never married",
                 77: "Refused", 99: "Don't know"},
    "RIDEXPRG": {1: "Yes, positive lab pregnancy test", 2: "Not pregnant",
                 3: "Cannot ascertain"},
    "DMDHRGND": {1: "Male", 2: "Female"},
    # ALQ_L : alcohol
    "ALQ111": YESNO,
    "ALQ121": {0: "Never in the last year", 1: "Every day", 2: "Nearly every day",
               3: "3 to 4 times a week", 4: "2 times a week", 5: "Once a week",
               6: "2 to 3 times a month", 7: "Once a month",
               8: "7 to 11 times in the last year", 9: "3 to 6 times in the last year",
               10: "1 to 2 times in the last year", 77: "Refused", 99: "Don't know"},
    "ALQ151": YESNO,
    # SMQ_L : smoking
    "SMQ020": YESNO,
    "SMQ040": {1: "Every day", 2: "Some days", 3: "Not at all",
               7: "Refused", 9: "Don't know"},
    "SMAQUEX2": {1: "CAPI (interviewer)", 2: "ACASI (self)"},
    # PAQ_L : physical-activity unit fields (letter codes)
    "PAD790U": {"D": "Day", "W": "Week", "M": "Month", "Y": "Year"},
    "PAD810U": {"D": "Day", "W": "Week", "M": "Month", "Y": "Year"},
    # HSQ_L : health status
    "HSQ590": YESNO,
    # INQ_L : income
    "INDFMMPC": {1: "Index <= 1.30", 2: "1.30 < Index <= 1.85", 3: "Index > 1.85",
                 7: "Refused", 9: "Don't know"},
    "INQ300": YESNO,
}
```

#### Decoding the cells

```python
def decode(df):
    out = df.copy()
    for col in out.columns:
        if col in VALUE_LABELS:                        # only mapped categoricals
            m = VALUE_LABELS[col]
            out[col] = out[col].map(lambda v, m=m: m.get(v, v))   # keep if unmapped
    return out
```

#### Build codebook, decode all, merge, save one xlsx

```python
codebook, decoded_frames, merged = [], {}, None
for name, df in data.items():
    meta  = metas[name]
    topic = FILE_TOPICS.get(name, name)
    for col in df.columns:
        q = meta.column_names_to_labels.get(col, "")
        if col in VALUE_LABELS:
            for code, meaning in VALUE_LABELS[col].items():
                codebook.append([topic, name, col, q, code, meaning])
        else:
            kind = "respondent ID" if col == "SEQN" else "continuous / numeric"
            codebook.append([topic, name, col, q, "(all values)", kind])

    decoded_frames[topic[:31]] = decode(df)            # sheet name <= 31 chars
    merged = df if merged is None else merged.merge(df, on="SEQN", how="outer")

codebook_df = pd.DataFrame(codebook,
    columns=["Topic", "File", "Variable", "Question", "Code", "Meaning"])

with pd.ExcelWriter(MAPPED_XLSX, engine="openpyxl") as xl:
    codebook_df.to_excel(xl, sheet_name="Codebook", index=False)
    for sheet, dec in decoded_frames.items():
        dec.to_excel(xl, sheet_name=sheet, index=False)
    merged.to_excel(xl, sheet_name="Merged_Decoded", index=False)

print(f"Saved {MAPPED_XLSX} | codebook rows={len(codebook_df)} | merged={merged.shape}")
```
